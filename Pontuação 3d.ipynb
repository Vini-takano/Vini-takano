```python
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits import mplot3d

from matplotlib import animation
from matplotlib.animation import PillowWriter

plt.style.use(['default', 'notebook', 'grid'])

x_1= np.linspace(12.30,13.10,50) #CP
y_1 = np.linspace(2.08,2.28,50) #PV

x,y = np.meshgrid(x_1,y_1)

z = 10*(1.5/(1+np.exp((1.875-x/y)/1.35 + 0.2)) - 0.5)*x  #Fator de eficiência estrutural

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(x,y,z, cmap='plasma', linewidth=0, antialiased=False)
ax.view_init(elev=10, azim=0)

ax.set_title("Pontuação em função de CP e PV")
ax.set_xlabel("CP")
ax.set_ylabel("PV")
ax.set_zlabel("P")

def animate(i):
    ax.view_init(elev=10, azim=1*i)
    
ani = animation.FuncAnimation(fig, animate, frames=360, interval=50)
ani.save('teste.gif',writer='pillow',fps=30,dpi=100)
```

```python
plt.style.use(['science', 'notebook', 'grid'])

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

ax.plot_surface(x,y,z, cmap='plasma', linewidth=0, antialiased=False)
ax.view_init(elev=30, azim=45)

ax.set_title("Pontuação em função de CP e PV")
ax.set_xlabel("CP")
ax.set_ylabel("PV")
ax.set_zlabel("P")
```
