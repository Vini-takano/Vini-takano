```python
import numpy as np
import matplotlib.pyplot as plt

def function_p(x, y):
    return 10*(1.5/(1+np.exp((1.875-x/y)/1.35 + 0.2)) - 0.5)*x
```

```python
passo = 50

x_values= np.linspace(12.5,14,passo) #CP
y_values = np.linspace(2.0,2.5,passo) #PV

x,y = np.meshgrid(x_values, y_values)
```

```python
z = function_p(x, y)
z.shape
```

```python
plt.style.use(['science', 'notebook', 'grid'])
fig, ax = plt.subplots(figsize = (8,6))
cs = plt.contour(x,y,z, levels = 25, cmap='plasma')
plt.clabel(cs, fontsize = 8)
plt.xlabel('CP [Kg]')
plt.ylabel('PV [Kg]')
```
