```python
import numpy as np
import matplotlib.pyplot as plt

def function_p(x, y):
    return 10*(1.5/(1+np.exp((1.875-(y-x)/x)/1.35 + 0.2)) - 0.5)*(y-x)

#y = MTOW (Maximun take-off weight)
#x = PV (Peso vazio)
```

```python
passo = 50

CP_values= np.linspace(12.5,14,passo) #CP
PV_values = np.linspace(2.0,2.5,passo) #PV

MTOW = CP_values + PV_values

x,y = np.meshgrid(PV_values, MTOW)
```

```python
z = function_p(x,y)
z.shape
```

```python
plt.style.use(['science', 'notebook', 'grid'])
fig, ax = plt.subplots(figsize = (8,12))
cs = plt.contour(x,y,z, levels = 25, colors='black')
plt.clabel(cs, fontsize = 8)
plt.xlabel('PV[Kg]')
plt.ylabel('MTOW [Kg]')
```
